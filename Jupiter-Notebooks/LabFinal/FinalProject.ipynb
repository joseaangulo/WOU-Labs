<p>It wasn’t until the late 19th century breakfast cereal was introduced. The first pioneers were John Harvey Kellogg a physician and his competitor C.W. Post. The root for breakfast cereal was based on a theory of gut health rather than an energy efficient breakfast. In the article it says, “both believed that the path to health and happiness ran through the digestive tract.” Kellogg would say, "The causes of indigestion are responsible for more deaths than all other causes combined." Post beat Kellogg and came out with his Grape Nuts which originally made with “molasses and maltose from barley flour, but no cane or beet sugar.” Kellogg’s first cornflakes were also sugar free at first. </p>
<p>While Kellogg was away in Europe in 1902, he left his younger brother, W.K. in charge of the development process. He added sugar to the cornflakes to improve its flavor. When Kellogg returned, he was outraged at his brother saying “he felt that sugar was unhealthy and argued vehemently against using it.” Unfortunately, consumers disagreed and the sugar stayed. This opened the market and allowed for more cereals adding more sugar throughout the years.</p>
<p>Many physicians disagree with the amount of sugar added to cereal. Dr. Anna Robins, a lecture in nutrition,  voiced her opinion on Kellogg’s most recent effort on reducing the amount of sugar in Coco Pops by 40% a lecture in nutrition as she told the BBC’s Wake up to Money program: ”Any move to be helping the general public to make healthier choices is a good one. But I don’t think they’re going far enough to be making these cereals a healthy option in the morning.”</p>
<p>The government recommends that children aged between seven and 10 should eat no more than 24g of sugar daily. With many cereals give half or more of that to their consumers especially those targeting children.</p>

<p>The mean in sugar is 5g which shows that all the cereal brands average about 7g of sugar or 30% of our suggested daily intake. It also says more than half have 7g of sugar and a quarter have more than 11g. These would add up to almost 50% of our daily intake.<p>

```python
sugars_rating = df[['calories', 'sugars']]
```

```python
import matplotlib.pyplot as plot
ax = sugars_rating.plot.bar(x="calories", y="sugars",rot=0)
ax.set_ylabel('Sugar')
ax.set_xlabel('Calories')
ax.set_title('Sugar added based on total calories')
```

<p>This bar chart was used to try and associate the amount of sugar to the total calories of the cereal. There appears to be no correlation. </p>
